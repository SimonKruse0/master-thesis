```python
def get_rows_for_group(group):
    functions = [f"f{i+1}" for i in range(24)]
    if group == 1:
        functions = [f"f{i+1}" for i in range(5)]
    if group == 2:
        functions = [f"f{i+1}" for i in range(5,9)]
    if group == 3:
        functions = [f"f{i+1}" for i in range(9,14)]
    if group == 4:
        functions = [f"f{i+1}" for i in range(14,19)]
    if group == 5:
        functions = [f"f{i+1}" for i in range(19,24)]
    rows = [functions]
    regmodels = ["BO_numpy","BO_BOH", "BO_Gaus", "BO_Naiv","BO_em"]
    for reg_model in regmodels:
        mask = [reg_model in x for x in df.index.get_level_values(0).values]
        mean_opt = df[mask].loc[:,"DIM2","f"].mean()[functions]
        mean_opt_evals = df[mask].loc[:,"DIM2","f_evals"].mean()[functions]
        rows.append([reg_model]+[f"{x:0.2f}({y:0.1f})" for (x,y) in zip(mean_opt,mean_opt_evals)])
    return rows
```

```python
import latextable
from texttable import Texttable
from tabulate import tabulate
```

```python
def generate_latex_table(group):
    rows = get_rows_for_group(group)

    texdoc = tabulate(rows, headers='firstrow', tablefmt='latex')
    path = "/home/simon/Documents/MasterThesis/master-thesis/thesis/Tables"
    with open(f'{path}/best_minimizer_group_{group}.tex', 'w') as fout:
        for i in range(len(texdoc)):
            fout.write(texdoc[i])
```

```python
for i in range(1,6):
    generate_latex_table(i)
```
