```python
import latextable
from texttable import Texttable
from tabulate import tabulate
```

```python
table_1 = Texttable()
table_1.set_cols_align(["l", "r", "c"])
table_1.set_cols_valign(["t", "m", "b"])
#table.set_cols_align(["c"] * 4)
rows = [["f1", "f2", "f3"],
                [f"{x:0.2f}" for x in mean_opt],
                [f"{x:0.1f}" for x in mean_opt],
                [f"{x:0.0f}" for x in mean_opt]]
table_1.set_deco(Texttable.HEADER | Texttable.VLINES)
table_1.add_rows(rows)

texdoc = tabulate(rows, headers='firstrow', tablefmt='latex')
#texdoc = latextable.draw_latex(table_1, caption="An example table.", label="table:example_table")
path = "/home/simon/Documents/MasterThesis/master-thesis/thesis/Tables"
with open(f'{path}/test.tex', 'w') as fout:
    for i in range(len(texdoc)):
        fout.write(texdoc[i])
```
